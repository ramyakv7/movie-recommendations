```python
df_filtered = df_new[(df_new['runtime'] >= 60) & (df_new['runtime'] <= 300)]
```

```python
df_filtered = df_filtered[df_filtered['vote_count'] >=160]
```

```python
df_filtered
```

```python
def weightedRating(X, m = 160, c = 5.6):
    v = X['vote_count']
    r = X['vote_average']
    ## compute weighted score
    wt_score = (v/(v+m) * r) + (m/(m+v) * c)
    ##            
    return wt_score


    
```

```python
wt_score = weightedRating(df_filtered)

wt_score
```

```python
df_filtered.apply(weightedRating, axis = 1)
```

```python
df_filtered['weighted_score'] = df_filtered.apply(weightedRating, axis = 1)
```

```python
df_filtered.head()
```

```python
top_10_movies = df_filtered.sort_values('weighted_score', ascending = False).head(10)
```
